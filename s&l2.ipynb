```python
lst=[i for i in range(1,20)]

print(lst)


def biAlg():
    low=0
    high=len(lst)
    flag=0
    num=int(input("number to be searched"))
    
    while flag==0:
        mid=(low+high)//2
        if num in lst:
            if mid==num:
                flag+=1
            elif mid<num:
                low=mid+1
            elif mid>num:
                high=mid+1
        else:
            flag+=1
            return "value not present"
        print("you searched{},is present in list at index{}".format(num,lst.index(num)))
biAlg()
```
